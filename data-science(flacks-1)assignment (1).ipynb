# Question-1 What is Flask Framework? What are the advantages of Flask Framework?

Answer-1-
Flask is a lightweight and flexible Python web framework that allows developers to build web applications quickly and easily. It is designed to be simple and easy to use while providing the essentials needed for building web applications. Flask is considered a microframework, meaning it is minimalistic and focused on providing core functionality, allowing developers to extend it with additional libraries and components as needed

# Here are some key advantages of using the Flask framework:

# Simplicity and Minimalism:
Flask is known for its simplicity and minimalistic design. It provides the basic features required to build a web application without imposing too much structure or complexity.

# Easy to Learn and Use:
Flask has a straightforward and intuitive API, making it easy for developers, especially beginners, to quickly grasp and start building applications. Its minimalistic design encourages rapid development.

# Flexibility:
Flask is highly flexible, allowing developers to choose and integrate the specific libraries and tools they need for their project. This flexibility enables customization and tailoring to project requirements.

# Modularity:
Flask is designed in a way that allows developers to choose and use components (such as ORM, authentication, caching) as needed. This modularity enables the creation of scalable and maintainable applications.

# Extensible:
Flask provides a wide range of extensions and plugins that can be easily integrated to add additional features and functionalities to the application, saving development time and effort.

# Question-3. What is App routing in Flask? Why do we use app routes?

Answer-3-In Flask, app routing refers to the mechanism of associating URLs (Uniform Resource Locators) with specific functions or views within a web application. This association is established using the @app.route() decorator, which allows you to map a URL to a Python function in your Flask application.

# Why do we use app routes?
# URL Mapping:
App routes allow you to map specific URLs (e.g., '/', '/about') to designated functions within your Flask application. This helps organize and structure the application, making it easier to handle different endpoints.

# Request Handling:
When a user accesses a particular URL associated with an app route, the corresponding function is called to handle the request. This allows for dynamic content generation based on the URL accessed by the user.

# Modular Code Organization:
Using app routes encourages a modular approach to code organization. Each route and its corresponding function can handle a specific aspect or feature of the application, promoting a cleaner and more maintainable codebase.

# Separation of Concerns:
App routes facilitate the separation of concerns by allowing developers to assign specific functions to handle different routes. This separation improves code readability, reusability, and maintainability.

# Flexibility and Scalability:
App routes provide the flexibility to scale your application by easily adding or modifying routes as needed. This enables the integration of new features or endpoints without disrupting the existing functionality.
